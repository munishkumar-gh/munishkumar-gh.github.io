### Functions

```python
# Functin for one hot encoding
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res) 
```

```python
mod_df = df_drop.copy()

features_to_encode = ['10_Region', '11_Basin name', '72_Onshore or offshore',
                      '82_Drive mechanism (main)', '149_Reservoir period', '153_Depositional system (main)',
                      '161_Reservoir architecture (main)', 
                     ]
for feature in features_to_encode:
    mod_df = encode_and_bind(mod_df, feature)
```

### 1. Train-Test split

```python
train_dataset = mod_df.sample(frac=0.8, random_state=42)
test_dataset = mod_df.drop(train_dataset.index)
```

```python
train_features = train_dataset.copy()
test_features = test_dataset.copy()

print('Training Set - Size:', train_features.shape)
print('Testing Set - Size:', test_features.shape)
```

```python
train_labels = train_features.pop('307_Recovery factor (ultimate oil) (%)')
test_labels = test_features.pop('307_Recovery factor (ultimate oil) (%)')

print('Predictor For Training Set - Size:', train_labels.shape)
print('Predictor For Testing Set - Size:', test_labels.shape)
```

### 2. Normalisation

## Multivariant Linear Analysis

```python
linear_model = tf.keras.Sequential([
    normalizer,
    layers.Dense(units=1)
])
```

```python
linear_model.layers[1].kernel
```

```python
linear_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_squared_error')
```

```python
%%time
history = linear_model.fit(
    train_features,
    train_labels,
    epochs=100,
    # Suppress logging.
    verbose=0,
    # Calculate validation results on 20% of the training data.
    validation_split = 0.2)
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  #plt.ylim([0, 10])
  plt.xlabel('Epoch')
  plt.ylabel('Error [RF]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```

```python
test_results = {}

test_results['linear_model'] = linear_model.evaluate(
    test_features, test_labels, verbose=0)
```

## Deep Nueral Network

```python
def build_and_compile_model(norm, hidden_layer_size, activation):
  model = keras.Sequential([
      norm,
      layers.Dense(hidden_layer_size, activation=activation),
      layers.Dense(hidden_layer_size, activation=activation),
      layers.Dense(hidden_layer_size, activation=activation),
      layers.Dense(hidden_layer_size, activation=activation),
      layers.Dense(1)
  ])

  model.compile(
      optimizer='adam',
      #optimizer=tf.keras.optimizers.Adam(0.01),
      #loss='sparse_categorical_crossentropy', #classification scenario, not applicable here
      loss = 'mean_squared_error',
      #metrics = ['accuracy']                
                )
  return model
```

```python
## Parameters to change
hidden_layer_size = 30
activation = 'relu'
batch_size = 50
max_epochs = 100
early_stopping = keras.callbacks.EarlyStopping(monitor = 'val_loss', patience =3)

dnn_model = build_and_compile_model(normalizer, hidden_layer_size, activation)
```

```python
%%time

history = dnn_model.fit(
    train_features,
    train_labels,
    batch_size = batch_size,
    epochs = max_epochs,
    callbacks = [early_stopping],
    # Suppress logging
    #verbose = 0,
    # Logging
    verbose=2,
    # Calculate validation results on 20% of the training data.
    validation_split = 0.2,
    validation_steps = 1
)
```

```python
dnn_model.summary()
```

```python
plot_loss(history)
```

```python
test_results['dnn_model'] = dnn_model.evaluate(test_features, test_labels, verbose=0)
```
