```python
import itertools
import snscrape.modules.twitter as sntwitter
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Collection
Scraping Twitter data

### Tweets from USA

Renewable Energy Tweets

```python
## USA Renewable Energy Tweets
# Creating list to append scraped tweet data
tweets_list_re_usa = []

# Search Filters
max_tweets = 10000
search_text = '(renewable energy) OR (#renewableenergy)'
tweets_since = '2021-06-01'
tweets_until = '2022-06-01'
lang = 'en'
loc_near = 'USA'

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('{0} since:{1} until:{2} lang:{3} near:{4}'.format(search_text, tweets_since, tweets_until, lang, loc_near)).get_items()):
    if i > max_tweets:
        break
    tweets_list_re_usa.append([tweet.date, tweet.rawContent, tweet.hashtags, tweet.lang, tweet.user.location, tweet.coordinates, tweet.place])
    
# Creating a dataframe from the tweets list above
tweets_df_re_usa = pd.DataFrame(tweets_list_re_usa, columns=['date_time', 'content', 'hashtags', 'language', 'user_location', 'coord', 'place'])
tweets_df_re_usa
```
