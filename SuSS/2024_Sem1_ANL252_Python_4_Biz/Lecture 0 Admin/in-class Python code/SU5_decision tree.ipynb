```python
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the iris dataset into pandas DataFrame
from sklearn.datasets import load_iris
iris_dataset = load_iris()
df = pd.DataFrame(data=iris_dataset.data, columns=iris_dataset.feature_names)
df['target'] = iris_dataset.target
df.head()
```

```python
# Extract the dependent and independent variables
X = df.drop('target', axis=1)
y = df['target']
```

```python
# Create a decision tree classifier
dtc = tree.DecisionTreeClassifier(random_state=123)
```

```python
# Fit the decision tree
dtc.fit(X, y)
```

```python
# from sklearn import plot_tree
# control the plotting size 
plt.figure(figsize=(10, 8))
# Plot the decision tree
tree.plot_tree(dtc, filled=True, feature_names=X.columns, class_names=iris_dataset.target_names)#,max_depth=2
```
