3rd activity

```python
car_model = pd.read_csv('./car_model.csv', na_values='na_string', na_filter=True)
car_price = pd.read_csv('./car_price.csv', na_values='na_string', na_filter=True)
car_price_model = car_model.merge(car_price, how='inner', on=['Year', 'Make', 'Model'])
car_price_model.dropna(axis=0, how='any')
```

```python
car_price_model.head()
```

4th activity

```python
clean = car_price_model.dropna(subset=['Price'])
```

```python
#confirm no missing values
clean['Price'].isnull()
```

```python
# get the 25th percentile, and the 75th percentile
q1 = clean['Price'].quantile(.25)
q3 = clean['Price'].quantile(.75)
```

```python
# get the interquartile range
iqr = q3 - q1
```

```python
# not considered outliers
clean[ ~( (clean['Price']< q1 - 1.5*iqr) | (clean['Price']> q3 + 1.5*iqr) ) ]
```

```python
# considered outliers
clean[( (clean['Price']< q1 - 1.5*iqr) | (clean['Price']> q3 + 1.5*iqr) )]
```
