```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
iris = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
iris.head()
```

```python
iris['Class'] = iris_dataset.target
```

```python
#check for missing values
iris.isnull().sum()
```

```python
# Extracting the dependent and independent variables
#X = pd.DataFrame(np.c_[boston['LSTAT'], boston['RM']], columns = ['LSTAT','RM'])
#X = data[:, [1, 9]]
#X = pd.DataFrame(iris['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
#X = pd.DataFrame(iris[['sepal length (cm)'],['sepal width (cm)']])
#X = iris['sepal length (cm)']
X =  pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
Y = iris['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn import tree
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import classification_report

# a classifier works for categorical target values
dtc = tree.DecisionTreeClassifier()
```

```python
# fit decision tree on training data
dtc.fit(X_train, Y_train)
# then use the fitted decision tree to predict on test data
# if decision tree learnt the true relationships, it should do well on test data
Y_predict = dtc.predict(X_test)
```

```python
# also, print classification report
print(classification_report(Y_predict,Y_test))
```

```python
# from sklearn import plot_tree
# control the plotting size 
plt.figure(figsize=(15, 8))
# plot the decision tree
tree.plot_tree(dtc)
plt.show()

# if you encounter error with plot_tree, try the updates in the next cell
```
