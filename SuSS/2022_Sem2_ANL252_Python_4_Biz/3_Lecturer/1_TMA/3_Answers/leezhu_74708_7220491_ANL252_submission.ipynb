```python
import numpy as np
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
from statistics import mean


tma = pd.read_csv(r"TMA_Data.csv")

staff = np.array(tma.Staff)
joinDate = np.array(tma.JoinDate)
leftDate = np.array(tma.LeftDate)
results = []

may_1_2022 = datetime.datetime(2022, 5, 1)

for l,s in zip(leftDate, joinDate):
    s = datetime.datetime.strptime(s, '%m/%d/%Y')
    if not isinstance(l, str):
        l = may_1_2022
    else:
        l = datetime.datetime.strptime(l, '%m/%d/%Y')
    difference_in_years = round((l -s).days / 365,1)
    # print(difference_in_years)
    results.append(difference_in_years)
    
#Average Length of Service
avg = mean(results)
print("The average is ", round(avg,1))

#Max Length of Service
print("\nMax Length of Service - " + staff[results.index(max(results))] + ":" + str(max(results)) + "\n")

#Min Length of Service
print("Min Length of Service - " + staff[results.index(min(results))] + ":" + str(min(results)) + "\n")

length_of_service = "\n".join("{} {}".format(x, y) for x, y in zip(staff, results))    
print(length_of_service)
```

```python
import numpy as np
import pandas as pd

tma = pd.read_csv(r"TMA_Data.csv")

staff = np.array(tma.Staff)
while(True):
    name = input("Enter a name to check the records (type 'exit' to stop finding): ")
    if name.casefold() == "exit".casefold():
        break
    elif name.casefold() in map(str.casefold, staff): 
        print(name + " can be found in the company's records. \n")
    else:
        print(name + " does not exist company's records. \n")
```
