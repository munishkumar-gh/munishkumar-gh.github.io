# Study Unit 1
## We will begin by using the Boston housing data from sklearn libary. You will learn more about this library later.

```python
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
# print the data content. Note: the three dots in the middle of the output means values in the middle of each row are omitted

print(data)
```

```python
# print the target content, which is the MEDV field (Median value of owner-occupied homes in $1000's)
# Target contains a set of values used for predictive modelling, which we will not worry about for now.
print(target)
```

```python
# the target contains a set of values (i.e., median house prices) stored in a data format called "array"
# you will learn more about arracy later
# here are some simple operations we can do about the access
# assign the values housing.target to another array named housing_price"
housing_price = target
```

```python
# we can print the content of the first entry in the array
print(housing_price[0])
```

```python
# we can print the first ten values of the median house prices
print(housing_price[0:10])
```

```python
# we can also use a for loop to print the first ten values of the median house prices
for x in housing_price:
 print(x)
```

```python
# we can also use a for loop to print the first ten values of the median house prices
for x in housing_price:
 print(x)
 if x == 18.9:
        break
```

```python
# Previous example is not elegant in two ways:
# You need to know that the 10th value is 18.9
# It is generally not a good practice to use break unless there is no other option
# following is a more elegant code to print the first ten values of the median house prices
for x in housing_price[0:10]:
 print(x)
 
```

```python
# Let's find the average price of the first ten vmedian house prices
avg = 0
for x in housing_price[0:10]:
 avg = avg + x
print("Average of first ten median house prices = ", avg/10)
```

```python
# you can do some plots using matplotlib
# we will cover matplotlib in details later
# we will just use it for now

import matplotlib.pyplot as plt

plt.boxplot(housing_price)
plt.ylabel('Housing Price')
plt.xlabel('Box Plot of Housing Price')
plt.show()
```

```python
# you can do some plots using matplotlib
# we will cover matplotlib in details later
# we will just use it for now

import matplotlib.pyplot as plt

# the histogram of the data
n, bins, patches = plt.hist(housing_price, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Housing Price')
plt.ylabel('Probability')
plt.title('Histogram of Housing Price')
plt.text(25, .04, r'$\mu=26.4$')
plt.xlim(0, 60)
plt.ylim(0, 0.1)
plt.grid(True)
plt.show()
```

```python
# the target contains a set of values (i.e., median house prices) stored in a data format called "array"
# you will learn more about array later
# here are some simple operations we can do about the access
# assign the values housing.target to another array named housing_price"
# find the sum, count, average values in the array, tuple, and list

# create a user-defined function
def average(total, number_of_items):
 return total/number_of_items

housing_price = target
print(sum(housing_price))
print(len(housing_price))
print(average(sum(housing_price),len(housing_price)))
print()

tuple_price = tuple(housing_price) #no need to use map since this is only one dimension
print(sum(tuple_price))
print(len(tuple_price))
print(average(sum(tuple_price),len(tuple_price)))
print()

list_price = housing_price.tolist()
print(sum(list_price))
print(len(list_price))
print(average(sum(list_price),len(list_price)))
print()
```

```python
# use data from Study Unit 2

import math
import numpy as numpy

housing_data = data
features_data = data
#for y in housing_data.ndim: 
#    print(housing_data[:,y])
print(housing_data.shape)
print(housing_data.size)
print(housing_data.ndim) #ndim = 2 because this is a 2D table
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

#compute the statistics for each column of the housing data
i = 0
while i < housing_data.shape[1]: 
 x = housing_data[:, i]
 print(feature_names[i], "\t\t min = ", numpy.min(x), \
       ", max =", numpy.max(x), ", std = ", round(numpy.std(x),3), \
       ", mean=", round(numpy.mean(x),3), ", median = ", round(numpy.median(x),3))
 i = i + 1
    
```

```python
import matplotlib.pyplot as plt

plt.bar(feature_names, housing_data.mean(axis=0), color ='maroon',
        width = 0.4)
 
plt.xlabel("Housing features")
plt.ylabel("Mean values")
plt.title("Mean values in different housing features")
plt.show()
```
