### Car sales program:

#### Read in the data file “car_outlier.csv” into Python as a pandas DataFrame. The data file contains 5 variables: Year, Make, Model, Category and Price.

#### Print the DataFrame and check on its dimension.
#### Sort the data by “make”, “model” and “year” in the ascending order entirely.

```python
import pandas as pd

car_outlier = pd.read_csv("car_outlier.csv")
car_outlier
```

```python
car_outlier.sort_values(
    by = ['Make', "Model", "Year"], ascending =True   
)
```

```python
car_outlier = car_outlier.rename(columns = {'Price':'Price (USD)'})
car_outlier
```

##### Since the prices stored in the “Price” variable are in US dollars, add a new column named “SGD Price” that is 1.5 times of the “Price”

##### Calculate the mean price (in SGD) of each car brand.

```python
 # Lets assume the exchange rate is USD 1 = SGD 1.5

ex_rate = 1.5

car_outlier['Price (SGD)'] = car_outlier['Price (USD)']*ex_rate
```

```python
car_outlier
```

```python
car_outlier.groupby("Category").mean()
```
