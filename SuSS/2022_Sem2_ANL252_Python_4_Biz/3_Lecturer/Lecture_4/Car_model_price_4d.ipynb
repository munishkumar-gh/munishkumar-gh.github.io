##### Remove all the rows with missing data in “price” from the merged DataFrame created in the previous chapter. 

```python
import pandas as pd

car_model_price = pd.read_csv("car_model_price.csv")
car_model_price
```

```python
car_model_price_clean = car_model_price.dropna(axis = 0, how = 'any')
car_model_price_clean
```

##### Detect car models that can be considered as outliers in terms of their selling price.

##### Discuss whether the models with extraordinarily high/low selling prices should be excluded from the DataFrame.

```python
q1 = car_model_price_clean['Price'].quantile(q = 0.25)
q3 = car_model_price_clean['Price'].quantile(q = 0.75)
iqr = q3-q1

y1 = q1 - 1.5*iqr
y2 = q3 + 1.5*iqr

print(f"Lower Bound:\n {y1}")
print(f"Upper Bound:\n {y2}")
```

```python
car_outlier = car_model_price_clean[~( 
    (car_model_price_clean["Price"]< y1) | (car_model_price_clean["Price"]> y2)
)]

car_outlier
```

```python
car_outlier.to_csv("car_outlier.csv", index = False)
```
