```python
car_model = pd.read_csv('./car_model.csv', na_values='na_string', na_filter=True)
car_model.head(5)
```

```python
car_price_model = pd.read_csv('./car_price_model.csv', na_values='na_string', na_filter=True)
car_price_model.head(5)
```

```python
# Join the two datasets using inner joins
car_joined_model = car_model.merge(car_price_model, how='inner', on=['Year', 'Make', 'Model'])
#car_joined_model.dropna(axis=0, how='any')
car_joined_model
```

# Fourth activity

```python
# check for nulls in the Price field
car_joined_model['Price'].isnull()
```

```python
clean = car_joined_model.dropna(subset=['Price'])
clean
```

```python
#confirm no missing values
clean['Price'].isnull()
```

```python
# get the 25th percentile, and the 75th percentile
q1 = clean['Price'].quantile(.25)
q3 = clean['Price'].quantile(.75)
```

```python
# get the interquartile range
iqr = q3 - q1
```

```python
# list statuses of all prices: True means not an oulier
# note: & is a bitwise AND operator
(clean['Price'] > q1 - 1.5*iqr) & (clean['Price'] < q3 + 1.5*iqr)
```

```python
# show records with normal prices (there are 216 records)
clean[(clean['Price'] > q1 - 1.5*iqr) & (clean['Price'] < q3 + 1.5*iqr)]
```

```python
# show records outside normal prices
# Python's Tilde ~n operator is the bitwise negation operator: 
# it takes the number n as binary number and “flips” all bits 0 to 1 and 1 to 0 to obtain the complement binary number.
# For example, the tilde operation ~1 becomes 0 and ~0 becomes 1 and ~101 becomes 010

clean[~((clean['Price'] > q1 - 1.5*iqr) & (clean['Price'] < q3 + 1.5*iqr))]
```

```python
import pandas as pd
car_price_model = pd.read_csv("./car_price_model.csv")
```

```python
car_price_model
```

```python
# Drop the rows where at least one element is missing.
car_price_model = car_price_model.dropna()
car_price_model
```

```python
# Add a new column named SGD Price 1.5 times of the Price
car_price_model['SGD Price'] = car_price_model['Price']*1.5   
car_price_model
```

```python
# compute the mean SGD price for each make

car_price_model.groupby('Make')['SGD Price'].mean()
```
