### <font color=black><i>Q2 - Conduct four (4) data pre-processing tasks for the analysis of the data, explaining results obtained.</i><font color=black>

```python
#------Q2-----

#to confirm the data types in columns read in the DataFrame
raw_df.dtypes
```

```python
#to confirm if there are any missing values present, and if yes how many and in which column
raw_df.isnull().sum()
```

```python
#a copy of raw_df is created for the purposes of cleaning the data
clean_df = raw_df.copy()
#as both Education and Marital are nominal data points, the missing values are replaced with the mode from each category
clean_df['EDUCATION'].fillna(clean_df['EDUCATION'].mode()[0], inplace=True)
clean_df['MARITAL'].fillna(clean_df['MARITAL'].mode()[0], inplace=True)
#this is to confirm that the above procedure have correctly been effected and there are no longer any missing values
clean_df.isnull().sum()
```

```python
#as both Age is a ratio data point, the invalid values are replaced with the median value
clean_df["AGE"].replace(to_replace=[-1,199], value=clean_df["AGE"].mode()[0], inplace=True)
```

```python
#this is to identify the invalid billing values in columns B1 to B5 (i.e. the values that are negative) via boolean clean_masking
B1_filt = (clean_df['B1'] < 0)
B2_filt = (clean_df['B2'] < 0)
B3_filt = (clean_df['B3'] < 0)
B4_filt = (clean_df['B4'] < 0)
B5_filt = (clean_df['B5'] < 0)


clean_df['B1'].loc[B1_filt] = clean_df['B1'].loc[~B1_filt].median()
clean_df['B2'].loc[B2_filt] = clean_df['B2'].loc[~B2_filt].median()
clean_df['B3'].loc[B3_filt] = clean_df['B3'].loc[~B3_filt].median()
clean_df['B4'].loc[B4_filt] = clean_df['B4'].loc[~B4_filt].median()
clean_df['B5'].loc[B5_filt] = clean_df['B5'].loc[~B5_filt].median()

clean_df
```

```python
#this is to replace the invalid value in the 'R3' columns, i.e. $0, with 0 instead
clean_df["R3"] = clean_df["R3"].str.replace('$', '', regex=True)
clean_df["R3"] = clean_df["R3"].str.replace(',', '', regex=True)
#this is go recast the 'R3' column as an integer
clean_df["R3"].astype(int)
clean_df
```

```python
#a function is defined to determine the skewness of distribution is defined
#this function takes 3 arguments - mode, mean and mean
    #if there is no mode value, the user will have to input 'no mode' as the argument for mode
def skewness(mode,median,mean):
    if not mode == 'no mode':
        if mode < median and median < mean:
            print("The data is positively skewed")
        elif mean < median and median < mode:
            print("The data is negatively skewed")
        elif mean == median and median == mode:
            print("The data is normally distributed")
    elif mode == 'no mode':
        if median < mean:
            print("The data is positively skewed")
        elif median < mode:
            print("The data is negatively skewed")
        elif  median == mode:
            print("The data is normally distributed")
```

### <font color=black><i>Q4 - Perform linear regression modelling to predict the variable, B1, explaining the approach taken, including any further data pre-processing.</i><font color=black>

```python
#------Q4-----

#the relationship between the various variables in the dataset is first explored to determine which should be included in the regression model
corr_df = clean_df.corr()
corr_df
```

```python
#as the S(n), B(n) and R(n) variables each have strong auto-correlations between the current value and its lagged values, the lagged values are dropped from the dataframe and excluded from the regression model
corr_df.drop(['B2','B3','B4','B5','S2','S3','S4','S5','R2','R4','R5'],axis='columns', inplace=True)
corr_df.drop(['B2','B3','B4','B5','S2','S3','S4','S5','R2','R4','R5'],axis='rows', inplace=True)
corr_df[corr_df>=0.75]
```

```python
#the dataset is first partitioned into both training and testing datasets.
#80% of the dataset will be used for training the model, and 20% for testing the model accuracy
#the variables that are auto-correlated or have multicollinearity (i.e. 'LIMIT' and 'INCOME')are not included in the regression model
x = clean_df[['LIMIT','BALANCE','RATING','GENDER','EDUCATION','MARITAL','AGE','S1','R1']]
y = clean_df['B1']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#this is to compute the p-values of all of the predictor variables
OLS(y_train,x_train).fit().summary()
```

```python
#the variables that have a p-values > 0.05 are dropped from the regression model
x.drop(['RATING','GENDER','EDUCATION','AGE','S1','R1'], axis='columns', inplace=True)

#as the x dataframe has changed, the training and testing data will have to be re-partioned
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#this is to compute the p-values of all of the remaining predictor variables
OLS(y_train,x_train).fit().summary()
```

```python
#this is to create an instance of the LinearRegression object as LM
lm =LinearRegression()

#this is to initialise the linear regression model using the training data
lm.fit(x_train,y_train)
```

### <font color=black><i>Q5 - State the linear regression equation and explain key insights from the results obtained in Question 4.</i><font color=black>

```python
#------Q5-----

#this is to obtain the coeffecients in the regression model
regression_coeff = lm.coef_
#this is to obtain the y-intercept in the regression model
regression_intercept = lm.intercept_
#the regression output is displayed in a formatted string for the sake of readability
print(f'B1 = {regression_coeff[0]}(LIMIT) + {regression_coeff[1]}(BALANCE) + {regression_coeff[2]}(MARITAL) + {regression_intercept}')
```
