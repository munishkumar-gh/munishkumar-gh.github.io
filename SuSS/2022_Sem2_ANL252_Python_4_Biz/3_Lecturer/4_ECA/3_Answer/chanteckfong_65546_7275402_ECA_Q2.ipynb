ii. Handling extremes/ outlier data

```python
df1.describe()
# Find the statistical summary from dataset
```

```python
# From the statistical summary above, age -1 and age 199 are the outliers.
```

```python
df1['AGE'].value_counts()[199]
# count the total number of occurrence of age "199" in the dataset
```

```python
df1['AGE'].value_counts()[-1]
# count the total number of occurrence of age "-1" in the dataset
```

```python
# Completely remove the 10 outliers' data points consider 'AGE' variable had a minimum of -1 and a maximum value of 199. 

index = df1[(df1['AGE'] >= 199) | (df1['AGE'] <= -1)].index # creates an index for all the data points where age takes these 2 values 
df1.drop(index, inplace=True)  # remove these index rows from data
df1['AGE'].describe()  # print summary statistics for the variable to ensure it has been removed 
```

iii. Data Discretisation 

```python
# Discretising a numeric variable into 8 bins
df1['AGE GROUP'] = pd.cut(df1['AGE'], bins = (0,20,30,40,50,60,70,80,100),right=False, labels =['Age <20','Age 20-29','Age 30-39','Age 40-49','Age 50-59','Age 60-69','Age 70-79','Age > 80'] )
display(df1)
```

```python
df1.drop(['ID'], axis=1, inplace=True) #reduce variable as ID column unnecessary
df1.head(2)
```
