```python
import pandas as pd
ship = pd.read_csv("ship.csv", na_values = ".", na_filter = True)
```

```python
rename_dict = {"T": "types", "A": "c_years", "P":
"o_periods", "MS": "s_months", "Y":
"incidents"}
ship.rename(columns = rename_dict, inplace = True)
ship
```

```python
shipgroup = round(ship[["types", "o_periods", "s_months",
"incidents"]].groupby(by = ["types",
"o_periods"]).mean(), 0)
shipgroup
```

```python
# Use a for-loop to iterate through all the ship types
for i in ship["types"].unique():
    # Use another for-loop to iterate through all the classes of operation periods
    for j in ship["o_periods"].unique():
        # Extract the mean service months of type i and operation j period from shipgroup
        rv_s_months = shipgroup["s_months"].loc[i].loc[j]
        
        # Extract the mean incidents of type i and operation j period from shipgroup
        rv_incidents = shipgroup["incidents"].loc[i].loc[j]
  
        # Create a Boolean mask for type i and operation period j with missing value in s_months 
        mask_s_months = (ship["types"] == i) & (ship["o_periods"] == j) & (ship["s_months"].isnull())
        
        # Create a Boolean mask for type i and operation period j with missing value in incidents
        mask_incidents = (ship["types"] == i) & (ship["o_periods"] == j) & (ship["incidents"].isnull())

        # Select the row of type i and op. period j with missing in s_months by the Boolean mask
        # Replace the value of it by the mean service months rv_s_months
        ship.loc[mask_s_months, "s_months"] = ship.loc[mask_s_months, "s_months"].fillna(rv_s_months)
        
        # Select the row of type i and op. period j with missing in incidents by the Boolean mask
        # Replace the value of it by the mean service months rv_incidents
        ship.loc[mask_incidents, "incidents"] = ship.loc[mask_incidents, "incidents"].fillna(rv_incidents)

# Print the result DataFrame
ship 
```

```python
Y = ship["incidents"]
Y
```

```python
ship = ship.astype({"types": "category", "c_years":
"category", "o_periods": "category"})
ship.dtypes
```

```python
x_ship = ship[["types", "c_years", "o_periods"]]
X = pd.get_dummies(x_ship)
X
```

```python
import numpy as np
X["log_s_months"] = np.log(ship["s_months"])
ship["log_s_months"] = np.log(ship["s_months"])
ship
```

```python
with open("ship_prepared.csv", "w") as write_to_file:
    for row in range(-1, len(ship)):
        if row == -1:
            line_content = list(ship.columns.values)
        else:
            line_content = [str(j) for j in ship.iloc[row]]
        line = ", ".join(line_content) + "\n"
        write_to_file.write(line)

write_to_file.close()


import sqlite3
conn = sqlite3.connect("ship.db")
cur = conn.cursor()
ship.to_sql("ship", conn, if_exists = "replace", index = False)
conn.commit()
conn.close()
```

```python
# Import linear_model module from sklearn
from sklearn import linear_model

# Initiate the PoissonRegressor estimator
clf = linear_model.PoissonRegressor()

# Fit the Poisson regression on the data X and Y
clf.fit(X, Y)

# Store the intercept as well as the coefficients in a list
# clf.intercept_ is used to extract the estimated intercept
# clf.coef_ is for the estimated coefficients
# Both attributes need to be converted to lists first
# After merging them into one list, the list will be converted to a NumPy array
coef = np.array([clf.intercept_] + list(clf.coef_))

# Create the labels by merging "intercept" (as a list) and the column labels of X
X_varnam = ["intercept"] + list(X.columns.values)

# Create a DataFrame with the coefficients as the data and the labels as the row indices
coef_table = pd.DataFrame(data = coef, index = X_varnam)

# Print the coefficient table
coef_table
```

```python
mu = clf.predict(X)
D = sum(2 * (Y * np.log(Y / mu, where = (Y > 0)) - (Y - mu)))
D
```
