## Learn about concating Tuples and Lists

```python
a = [1, 2, 3]
b = ['a', 'b', 'c']
```

```python
add_a = 4
list_new = a + [add_a]
print(list_new)
```

```python
add_a = [4]

list_new = a + add_a
print(list_new)
```

```python
b = ['a', 'b', 'c']
list_a = b + b
print(list_a)
```

```python
b_new = 'd'

list_a = b + [b_new]
print(list_a)
```

### This is a tuple

```python
tup_num1 = (1, 2, 3)
tup_alp1 = ('a', 'b', 'c')
```

```python
tup_num2 = 4
tup_alp2 = 'd'
```

```python
tup3 = tup_num1 + (tup_num2,)
print(tup3)
```

```python
tup4 = tup_alp1 + (tup_alp2,)
print(tup4)
```
