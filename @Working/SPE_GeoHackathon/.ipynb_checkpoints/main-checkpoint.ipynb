### 1. Start with uploading needed modules

```python
from model import Model
import pandas as pd
import matplotlib.pyplot as plt
from econ_simple import *
```

### 2. Initialize the model

<img style="float: left;" src="slides/Picture1.PNG" width="90%">

```python
# 60x40x40 grdi blocks with 30x30x2.5m size
m = Model()
m.add_well('INJ', 15 , 20)
m.add_well('PRD', 45 , 20)
m.init()
```

### 3. Add well controls and run model for 10 years

```python
for i, w in enumerate(m.reservoir.wells):
    if 'P' in w.name:
        w.control = m.physics.new_rate_water_prod(7000)
    elif 'I' in w.name:
        w.control = m.physics.new_rate_water_inj(7000, 273.15+25) #Kelvin

m.run_python(365*10)
```

### 4. Change well controls and run for another 2 years

```python
for i, w in enumerate(m.reservoir.wells):
    if 'P' in w.name:
        w.control = m.physics.new_rate_water_prod(5000)
    elif 'I' in w.name:
        w.control = m.physics.new_rate_water_inj(5000, 273.15+25) #rate cubic meters

m.run_python(365*2)
```

### 5. Export data and evaluate economy

```python
time_data = pd.DataFrame.from_dict(m.physics.engine.time_data) #well prod data
#writer = pd.ExcelWriter('time_data.xls')
#time_data.to_excel(write, 'Sheet1')
#write.save()
```

### 6. Plot well regimes and economics

```python
search_str = 'INJ : water rate' 
ax = time_data.plot(x='time', y=[col for col in time_data.columns if search_str in col])

search_str = 'INJ : BHP' 
ax = time_data.plot(x='time', y=[col for col in time_data.columns if search_str in col])

search_str = 'PRD : water rate' 
ax = time_data.plot(x='time', y=[col for col in time_data.columns if search_str in col])

search_str = 'PRD : temp' 
ax = time_data.plot(x='time', y=[col for col in time_data.columns if search_str in col])

plt.show()
```
